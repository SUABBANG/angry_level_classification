```python
import numpy as np
from scipy.io import wavfile

def max_decibel(wav_file_path):
    try:
        # WAV 파일을 읽어옵니다.
        sample_rate, audio_data = wavfile.read(wav_file_path)

        # 각 샘플의 제곱 값을 계산합니다.
        squared_audio_data = np.square(audio_data)

        # 각 샘플 제곱 값의 평균을 계산합니다.
        mean_squared = np.mean(squared_audio_data)

        # 평균 제곱 값의 루트를 씌워 최대 데시벨을 계산합니다.
        max_db = 10 * np.log10(mean_squared)

        return max_db
    except Exception as e:
        print(f"오류 발생: {e}")
        return None
```

```python
import librosa

def max_pitch(wav_file_path):
    try:
        # WAV 파일을 로드합니다.
        audio, sample_rate = librosa.load(wav_file_path, sr=None)

        # 피치 추출
        pitches, magnitudes = librosa.piptrack(y=audio, sr=sample_rate)

        # 각 프레임에서 최대 피치 값을 찾습니다.
        max_pitch_per_frame = pitches[magnitudes.argmax(axis=0)]

        # 모든 프레임의 최대 피치 값을 평균하여 최대 피치를 계산합니다.
        max_pitch_hz = max_pitch_per_frame.mean()

        return max_pitch_hz
    except Exception as e:
        print(f"오류 발생: {e}")
        return None
```

```python
for i in tqdm(range(len(angry_df))):
    audio_path = angry_df['dir'][i]
    print(audio_path)
    max_db = max_decibel(audio_path)
    max_pit = max_pitch(audio_path)  
    # average_speech_rate = average_speaking_rate(audio_path)
    
    angry_df['pitch'][i]=max_pit
    angry_df['decibel'][i]=max_db
    # angry_df['sps'][i]=average_speech_rate
```
