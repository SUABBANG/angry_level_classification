### 파일 가져오기

```python
# 폴더 내 모든 파일 불러오는 함수
import os

def list_files_in_directory(directory):
    file_list = []
    
    # 지정된 디렉토리의 모든 파일과 서브디렉토리를 순회합니다.
    for root, directories, files in os.walk(directory):
        for filename in files:
            # 파일의 전체 경로를 생성합니다.
            filepath = root + '/' +filename
            # filepath = os.path.join(root, filename)
            # 파일 리스트에 추가합니다.
            file_list.append(filepath)
    
    return file_list
```

```python
############# angry

# 폴더 경로를 지정합니다.
folder_path = 'C:/Users/user/DATA/voice/angry'  # 폴더 경로를 실제 경로로 바꿔주세요.

# 폴더 내의 파일을 리스트에 담습니다.
angry_files = list_files_in_directory(folder_path)
```

# 오디오 데이터셋

```python
import pandas as pd
import soundfile as sf
import librosa
import numpy as np
from tqdm import tqdm

# 'dir' 칼럼의 각 경로에 대해 오디오 파일을 열어 데이터를 읽고 딕셔너리를 생성합니다.
result = []
i = 0
for index, row in tqdm(dataset.iterrows()):
    file_path = row['dir']
    try:
        audio_data, sampling_rate = sf.read(file_path)
        # audio_data, sampling_rate = librosa.load(file_path, sr=16000)

        # # mono (1D) : 스테레오 타입 2차원으로 낭올 수 있음
        if len(audio_data.shape) == 2:
            audio_data = audio_data[:, 0]  # 첫 번째 채널 선택하여 모노로 바꿔줌
        
        audio_dict = {
            'path': file_path,
            'array': audio_data,
            'sampling_rate': sampling_rate,
            
        }
        result.append(audio_dict)
        i = i+1
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
```
